# Data Exploration Notebook

This notebook explores the fruit freshness dataset, including:
- Dataset statistics
- Class distribution
- Sample images
- Image characteristics


```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
import pandas as pd

# Set paths
data_dir = Path('../data')
train_dir = data_dir / 'train'
val_dir = data_dir / 'val'
test_dir = data_dir / 'test'

print("Dataset structure:")
print(f"Train: {train_dir}")
print(f"Val: {val_dir}")
print(f"Test: {test_dir}")
```

```python
# Count images in each class
def count_images(directory):
    counts = {}
    for class_name in ['fresh', 'rotten']:
        class_dir = directory / class_name
        if class_dir.exists():
            counts[class_name] = len(list(class_dir.glob('*.jpg')) + list(class_dir.glob('*.png')))
    return counts

train_counts = count_images(train_dir)
val_counts = count_images(val_dir)
test_counts = count_images(test_dir)

print("Training set:", train_counts)
print("Validation set:", val_counts)
print("Test set:", test_counts)
```

```python
# Visualize class distribution
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for idx, (name, counts) in enumerate([('Train', train_counts), ('Val', val_counts), ('Test', test_counts)]):
    axes[idx].bar(counts.keys(), counts.values())
    axes[idx].set_title(f'{name} Set Distribution')
    axes[idx].set_ylabel('Number of Images')

plt.tight_layout()
plt.show()
```

```python
# Display sample images
def show_samples(directory, num_samples=4):
    fig, axes = plt.subplots(2, num_samples, figsize=(15, 6))
    
    for class_idx, class_name in enumerate(['fresh', 'rotten']):
        class_dir = directory / class_name
        if class_dir.exists():
            images = list(class_dir.glob('*.jpg')) + list(class_dir.glob('*.png'))
            for img_idx in range(min(num_samples, len(images))):
                img = Image.open(images[img_idx])
                axes[class_idx, img_idx].imshow(img)
                axes[class_idx, img_idx].set_title(f'{class_name.capitalize()}')
                axes[class_idx, img_idx].axis('off')
    
    plt.tight_layout()
    plt.show()

show_samples(train_dir)
```
